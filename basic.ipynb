```python
# sognal reconstruct 
import numpy as np
from matplotlib import pyplot as plt

# Signal parameters

fs = 75  # sampling frequency
t = np.arange(0, 1, 1/fs)  # time vector
freq = 5  # signal frequency
amplitude = 1  # signal amplitude

# Signal generation

signal = amplitude * np.sin(2 * np.pi * freq * t)

# plot the signal 

plt.plot(t, signal)
plt.xlabel('Time (s)')
plt.ylabel('Signal amplitude')
plt.show()
```

```python
# how to create a damped_signal signal 

damping_factor = 0.05  # damping factor

damped_signal = signal * np.exp(-damping_factor * t)

# plot the damped signal

plt.plot(t, damped_signal)
plt.xlabel('Time (s)')
plt.ylabel('Damped signal amplitude')
plt.show()
```

```python
# Create a new sine wave with different parameters
import numpy as np 
from matplotlib import pyplot as plt 
new_freq = 10  # different frequency
new_amplitude = 2  # different amplitude
t = np.arange(0, 100, 1)  # create time vector from 0 to 100 with step 1
new_signal = new_amplitude * np.sin(2 * np.pi * new_freq * t)

# Plot the new sine wave
plt.plot(t, new_signal)
plt.xlabel('Time (s)')
plt.ylabel('Signal amplitude')
plt.title('New Sine Wave')
plt.grid(True)
plt.show()
```
